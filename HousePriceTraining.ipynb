```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
# Load data
X_train = pd.read_csv("X_train_scaled.csv")
X_test = pd.read_csv("X_test_scaled.csv")
y_train = pd.read_csv("y_train.csv")
y_test = pd.read_csv("y_test.csv")
```

```python
# Intialize MLP Model
mlp = MLPRegressor(
    hidden_layer_sizes=(128, 64),# hiddem layers
    activation='relu', # Activation Function learn complecated relationship
    solver='adam', #Optimezeer using weight
    max_iter=500,
    random_state=42
)
#Train
mlp.fit(X_train, y_train)
y_pred = mlp.predict(X_test)

# calculate Matrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("ü§ñ Sklearn MLP Neural Network Results")
print("-------------------------------------")
print(f"MAE  : {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R¬≤   : {r2:.4f}")
```

```python
# ============================================
# 2Ô∏è‚É£ MLPRegressor (Neural Network) Auto-Tuning
# ============================================
#define
mlp_param_grid = {
    'hidden_layer_sizes': [(64,), (128,64), (128,128), (256,128)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'sgd'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant', 'adaptive']
}
#intialize
mlp = MLPRegressor(max_iter=500, random_state=42)
#set up randomized search
mlp_search = RandomizedSearchCV(
    estimator=mlp,
    param_distributions=mlp_param_grid,
    n_iter=15, #number of configuration
    cv=3, #fold cross validation
    scoring='neg_mean_squared_error',
    random_state=42,
    verbose=2, #tuning process
    n_jobs=-1 #cpu core soeed
)
#train model search BP
mlp_search.fit(X_train, y_train)
#extract BEST MODEL
best_mlp = mlp_search.best_estimator_
# Prediction
y_pred_mlp = best_mlp.predict(X_test)

print("\nü§ñ Best MLP Params:", mlp_search.best_params_)
print(f"MLP MAE : {mean_absolute_error(y_test, y_pred_mlp):.4f}")
print(f"MLP RMSE: {np.sqrt(mean_squared_error(y_test, y_pred_mlp)):.4f}")
print(f"MLP R¬≤  : {r2_score(y_test, y_pred_mlp):.4f}")
```
